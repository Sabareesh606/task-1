```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample data
data = {
    'Age': [23, 25, 31, 22, 35, 29, 24, 26, 28, 32, 40, 23, 24, 31, 27],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Male', 'Female', 'Male', 'Female', 'Male', 'Male',
               'Female', 'Male', 'Female', 'Female', 'Male']
}

df = pd.DataFrame(data)

# Plot histogram for Age distribution
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.histplot(df['Age'], bins=5, kde=True, color='skyblue')
plt.title('Histogram: Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')

# Plot bar chart for Gender distribution
plt.subplot(1, 2, 2)
sns.countplot(x='Gender', data=df, palette='Set2')
plt.title('Bar Chart: Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```
