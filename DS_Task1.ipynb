```python
# Task 1
## importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn  as sns
```

```python
# Reading csv file
df=pd.read_csv("worldpopulationdata (1).csv")

#print only top 5 rows
df.head()
```

```python
#print only bottom 5 rows
df.tail()
```

```python
#checking the number of columns,information, shape and description about the dataset
df.columns
df.info()
df.shape
df.describe()
```

```python
#to check duplicates in dataset
df.duplicated().sum()
```

```python
#to check missing values
df.isnull().sum()
```

```python
#to check unique values in dataset
print(df['Country Name'].unique())
print("\n Total number of unique values:",df['Country Name'].nunique())
```

```python
#to check unique country code
print(df['Country Code'].unique())
print("\n Total number of unique values:",df['Country Code'].nunique())
```

```python
#to drop missing values
cols_to_drop = ['Series Name', 'Country Name']
df.drop(columns=[col for col in cols_to_drop if col in df.columns], inplace=True)
```

```python
#extracting top 10 countries by their total population

#filter data from total population
total_population=df[df['Series Code'] == 'SP.POP.TOTL']

#Sorting data based on the total population
total_population_sorted=total_population.sort_values(by="2022",ascending=False)

#print top 10 countries
total_top_ten_population=total_population_sorted.head(10)
print("\n Top 10 Countries of Highest Population\n")
print(total_top_ten_population['Country Code'])
```

```python
#create a bar plot for 2022
plt.figure(figsize=(15,8))
plt.subplot(2,2,1)
sns.barplot(x="2022", y="Country Code", hue="2022", data=total_top_ten_population, palette="Spectral", legend=False)
plt.title("Top Ten Countries of Total Population in(2022)",fontsize=12)
plt.xlabel("Total Population",fontsize=12)
plt.ylabel("Countries",fontsize=12)
plt.show()

#create a bar plot for 2016
plt.figure(figsize=(15,8))
plt.subplot(2,2,1)
sns.barplot(x="2016", y="Country Code", hue="2016", data=total_top_ten_population, palette="Spectral", legend=False)
plt.title("Top Ten Countries of Total Population in(2016)",fontsize=12)
plt.xlabel("Total Population",fontsize=12)
plt.ylabel("Countries",fontsize=12)
plt.show()
```

```python
#extracting bottom 10 countries by their total population

#Sorting data based on the total population
total_population_sorted1=total_population.sort_values(by="2022",ascending=True)

#print bottom 10 countries
total_bottom_ten_population=total_population_sorted1.head(10)
print("\n Bottom 10 Countries of Highest Population\n")
print(total_bottom_ten_population['Country Code'])
```

```python
#create a bar plot for 2022
plt.figure(figsize=(15,8))
plt.subplot(2,2,1)
sns.barplot(x="2022", y="Country Code", hue="2022", data=total_bottom_ten_population, palette="Spectral", legend=False)
plt.title("Bottom Ten Countries of Total Population in(2022)",fontsize=12)
plt.xlabel("Total Population",fontsize=12)
plt.ylabel("Countries",fontsize=12)
plt.show()

#create a bar plot for 2016
plt.figure(figsize=(15,8))
plt.subplot(2,2,1)
sns.barplot(x="2016", y="Country Code", hue="2016", data=total_bottom_ten_population, palette="Spectral", legend=False)
plt.title("Top Bottom Countries of Total Population in(2016)",fontsize=12)
plt.xlabel("Total Population",fontsize=12)
plt.ylabel("Countries",fontsize=12)
plt.show()
```

```python
#extracting top 10 countries by their highest male population

#filter data for male population
male_population=df[df['Series Code'] == 'SP.POP.TOTL.MA.IN']

#Sorting data based on the total male population
male_population_sorted=male_population.sort_values(by="2022",ascending=False)

#print top 10 countries of highest male population
male_top_ten_population=male_population_sorted.head(10)
print("\n Top 10 Countries of Highest Male Population\n")
print(male_top_ten_population['Country Code'])
```

```python
#extracting top 10 countries by their highest female population

#filter data for female population
female_population=df[df['Series Code'] == 'SP.POP.TOTL.FE.IN']

#Sorting data based on the total male population
female_population_sorted=female_population.sort_values(by="2022",ascending=False)

#print top 10 countries of highest male population
female_top_ten_population=female_population_sorted.head(10)
print("\n Top 10 Countries of Highest Female Population\n")
print(female_top_ten_population['Country Code'])
```

```python
#create a bar plot for 2022 with highest male and female population
plt.figure(figsize=(15,8))
plt.subplot(2,2,1)
sns.barplot(x="2022", y="Country Code", hue="2022", data=male_top_ten_population, palette="Spectral", legend=False)
plt.title("Top Ten Countries of Total male Population in(2022)",fontsize=12)
plt.xlabel("Total Population",fontsize=12)
plt.ylabel("Countries",fontsize=12)
plt.show()

#create a bar plot for 2016 with highest male and female population
plt.figure(figsize=(15,8))
plt.subplot(2,2,1)
sns.barplot(x="2016", y="Country Code", hue="2016", data=female_top_ten_population, palette="Spectral", legend=False)
plt.title("Top Ten Countries of Total female Population in(2016)",fontsize=12)
plt.xlabel("Total Population",fontsize=12)
plt.ylabel("Countries",fontsize=12)
plt.show()
```

```python
#top 5 countries with male and female population 

#merge male and female population data based on country code
merge_data=pd.merge(male_population,female_population,on="Country Code",suffixes=('_male','_female'))

#calculate the total population
merge_data['Total population']=merge_data['2022_male'] + merge_data['2022_female']

#print top 5 data
merge_data.head()
```

```python
#sorting data in descending order
sorted_data=merge_data.sort_values(by="Total population",ascending=False)

#top 10 countries with highest population
top_10_countries=sorted_data.head(10)
```

```python
#creating a stacked bar plot

plt.figure(figsize=(12,6))
sns.barplot(x="Country Code",y="2022_female",  data=top_10_countries, color="blue", label="female population")
sns.barplot(x="Country Code", y="2022_male",  data=top_10_countries, bottom=top_10_countries["2022_female"], color="yellow",  label="male population")
plt.xlabel("Countries")
plt.ylabel("Population")
plt.show()
```

```python
#bottom 10 countries
bottom_10_countries=sorted_data.tail(10)
```

```python
#creating a stacked bar plot

plt.figure(figsize=(12,6))
sns.barplot(x="Country Code",y="2022_female",  data=bottom_10_countries, color="blue", label="female population")
sns.barplot(x="Country Code", y="2022_male",  data=bottom_10_countries, bottom=bottom_10_countries["2022_female"], color="yellow",  label="male population")
plt.xlabel("Countries")
plt.ylabel("Population")
plt.show()
```
